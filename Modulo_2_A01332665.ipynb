```python
ejempli =[12,23,34,56,78 ,[23,23,]]


#Primero vamos a repasar como crear una tabla en PYTHON 




datos = [[10, 11, 12, 13],
         [20, 21, 22, 23],
         [30, 31, "" , 33]]

columnas = ['C1', 'C2', 'C3', 'C4'] # definimos los nombres de las columnas
filas = ['F1', 'F2', 'F3'] # definimos los nombres de las filas

df = pd.DataFrame(datos, columns=columnas, index=filas)

print(df)

#replacing data is better 

#replace variables with the average value 

#replace with frequency 

#leave the os




#missing variables ________________________

#Se utiliza la siguiente funcion: dataframes.dropna().


#df es la variable 

df.dropna(subset=["C3"],axis = 0, inplace = True)
print(df)

mean = df["C1"].mean()

print(mean)


df["C3"].replace(np.nan,mean)

print(df)


#Data formating ensure data consistent different ways that people write a word 

#we want to treat them as . Car Fuel we would like to transform the units 

#first we take the column we would like to work with 

df["C2"] = 2 * df["C2"]   #Modificamos el valor interno de cada columna 
df.rename(columns= {"C2":"C2 * 2"}, inplace = True )  #Cambiamos el valor de set 


#Cambiar el tipo de dato de la columna 

 

print(df)

```

```python
tablita = [[20,100000],[30,20000],[40,50000]]

columnas1 = ['EDAD', 'INCOME'] # definimos los nombres de las columnas
 # definimos los nombres de las filas
tablita1 = pd.DataFrame(tablita, columns = columnas1)

tablita1
```

```python
#Normalizacion_________________________________

#pordia ser con valores entre 0 y 1 



#Scaling.  lo divide entre el valor m'as grande en la columna 

tablita1['EDAD'] = tablita1['EDAD'] / tablita1['EDAD'].max()
tablita1['INCOME'] = tablita1['INCOME'] / tablita1['INCOME'].max()
tablita1
```

```python
#MAX MIN 

tablita1 = pd.DataFrame(tablita, columns = columnas1)
tablita1['EDAD'] = (tablita1['EDAD']- tablita1['EDAD'].min())/(tablita1['EDAD'].max()-tablita1['EDAD'].min())
tablita1['INCOME'] = (tablita1['INCOME']- tablita1['INCOME'].min())/(tablita1['INCOME'].max()-tablita1['INCOME'].min())
tablita1
```

```python
# Z CORE 

tablita1 = pd.DataFrame(tablita, columns = columnas1)

tablita1['EDAD'] = (tablita1['EDAD']- tablita1['EDAD'].mean())/tablita1['EDAD'].std()
tablita1['INCOME'] = (tablita1['INCOME']- tablita1['INCOME'].mean())/tablita1['INCOME'].std()
tablita1
```
